```python
import pandas as pd

# Create sample data
data = {
    'ID': range(1, 31),
    'Age': [23, 35, 45, 23, 31, 40, 23, 36, 38, 50, 27, 45, 29, 34, 50, 48, 28, 22, 33, 30, 37, 25, 29, 41, 43, 32, 46, 52, 39, 26]
}

# Create a DataFrame
df_population = pd.DataFrame(data)

# Specify file path with double backslashes or raw string
file_path = r'C:\PRODIGY_DS_01\population_data.csv'

# Save DataFrame to CSV file
df_population.to_csv(file_path, index=False)

# Print file path for confirmation
print(file_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  # Make sure to import numpy

# Load dataset (replace 'population_data.csv' with your actual file path)
df = pd.read_csv(r'C:\PRODIGY_DS_01\population_data.csv')

# Preview the dataset
df.head()

# Assuming 'Age' is the column with the age data
ages = df['Age']

# Create the histogram
plt.figure(figsize=(8, 6))
plt.hist(ages, bins=np.arange(ages.min(), ages.max() + 5, 5), edgecolor='black', color='skyblue')
plt.title('Age Distribution in Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
